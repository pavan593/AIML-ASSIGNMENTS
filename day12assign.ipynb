```python

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("banking_data.csv")




transaction_sum_by_account_type = df.groupby('Account_Type')['Transaction_Amount'].sum()


plt.figure(figsize=(8, 5))
transaction_sum_by_account_type.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Transaction Amount by Account Type', fontsize=14)
plt.xlabel('Account Type', fontsize=12)
plt.ylabel('Total Transaction Amount', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


transaction_count_by_branch = df['Branch'].value_counts()


plt.figure(figsize=(6, 6))
transaction_count_by_branch.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='tab10')
plt.title('Percentage of Transactions per Branch', fontsize=14)
plt.ylabel('')
plt.show()
```
