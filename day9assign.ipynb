```python
# Import necessary libraries
import pandas as pd

# Assignment 1: Data Exploration
# Load the banking_data.csv file
df = pd.read_csv("banking_data.csv")

# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())

# Generate basic statistics of the numerical columns
print("\nBasic statistics of numerical columns:")
print(df.describe())

# Check for missing values in the dataset
print("\nMissing values in the dataset:")
print(df.isnull().sum())

# Assignment 2: Data Aggregation and Grouping
# Group data by Account_Type and calculate the total sum of Transaction_Amount
transaction_sum_by_account_type = df.groupby('Account_Type')['Transaction_Amount'].sum()

# Calculate the average Account_Balance for each account type
average_balance_by_account_type = df.groupby('Account_Type')['Account_Balance'].mean()

# Group data by Branch and calculate the total number of transactions per branch
transaction_count_by_branch = df.groupby('Branch')['Transaction_Amount'].count()

# Calculate the average transaction amount per branch
average_transaction_by_branch = df.groupby('Branch')['Transaction_Amount'].mean()

# Display the results
print("\nTotal sum of Transaction_Amount by Account_Type:")
print(transaction_sum_by_account_type)

print("\nAverage Account_Balance by Account_Type:")
print(average_balance_by_account_type)

print("\nTotal number of transactions per Branch:")
print(transaction_count_by_branch)

print("\nAverage Transaction_Amount per Branch:")
print(average_transaction_by_branch)
```
