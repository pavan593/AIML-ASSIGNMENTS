ASSUMPTIONS IN MULTI LINEAR REGRESSION
1. linearity: The relationship between the predictors and the response is linear
2. Independence: Observations are independent of each other
3. Homoscedasticity: The residuals(difference between observed and predicted values) exhibit constant variance at all levels of the predictor
4. Normal Distribution of errors: The residuals of  the model are normally distributed.
5. No multicollinearity: The independent variable should not be too highly correlated with each other
