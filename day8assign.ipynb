```python
import pandas as pd

# Sample DataFrame
data = {
    "Region": ["East", "West", "East", "North", "South"],
    "Sales": [1500, 800, 1200, 400, 2000],
    "Profit": [300, 100, 400, 50, 500],
    "Quantity": [30, 10, 40, 5, 50],
}
df = pd.DataFrame(data)

# Data Filtering
# 1. Extract all rows where sales are greater than 1000
sales_greater_than_1000 = df[df["Sales"] > 1000]

# 2. Find all sales records for the region "East"
sales_in_east = df[df["Region"] == "East"]

# Data Processing
# 3. Add a new column, Profit_Per_Unit
df["Profit_Per_Unit"] = df["Profit"] / df["Quantity"]

# 4. Create another column, High_Sales, which labels rows as Yes if Sales > 1000, else No
df["High_Sales"] = df["Sales"].apply(lambda x: "Yes" if x > 1000 else "No")

# Display results
print("Rows with Sales > 1000:")
print(sales_greater_than_1000)

print("\nSales records for region 'East':")
print(sales_in_east)

print("\nDataFrame after processing:")
print(df)
```
